##시리즈
인덱스와 값으로 이루어진 1차원 형태 자료
+ 특별히 지정하지 않으면 인덱스는 보통의 연속형 자료들처럼 0부터 시작 
+ 인덱스를 지정하려면 원소의 개수가 동일한 자료를 함수 Series()의 index옵션으로 지정

ex)index를 1월, 2월..이런식으로 지정가능
+


```python
import pandas as pd
#인덱스 지정
myindex = ["1월", "2월","3월","4월","5월","6월"]
mydata = ["일월", "이월", "삼월", "사월", "오월", "유월"]
sr = pd.Series(mydata, index = myindex)
print(sr)

#지정한 인덱스를 이용해 자유롭게 인덱싱 가능
print(sr['1월'])



```

###데이터프레임
+ 행과 열이 있는 테이블 형태의 자료
+ 행 인덱스와 열 인덱스 쌍으로 각 요소 구분
+데이터프레임 생성하기 : 여러 개의 시리즈나 리스트를 이용하여 생성 : 각각의 시리즈 또는 리스트가 하나의 열이 됨 
+ 데이터프레임 요소 가져오기 
  + df.loc[행 인덱스, 열 인덱스] : 인덱스는 '시작:끝' 형태로 범위를 나타내거나 리스트 형태
  + df.iloc[행 순서, 열 순서] : 순서는 '시작:끝' 형태로 범위를 나타내거나 리스트 형태, 단 범위를 나타낼 시 '끝'에 해당하는 순서 미포함
  +df[조건식] : 조건식을 만족하는 요소만 출력

+ 데이터프레임 정보 살펴보기
  + df.head(n) : 상위 n개의 데이터 조회 (n의 default 값은 5)
  + df.tail(n) : 하위 n개의 데이터 조회 (n의 default 값은 5)
  + df.info() : 행과 열의 크기, 열의 이름, 열을 구성하는 값의 자료형 등을 출력
  + df.describe() : 열 별 데이터의 개수, 평균값, 표준편차, 최솟값, 최댓값, 4분위수 등 간단한 통계 정보
  + df.corr() : 수치 데이터를 가진 열 간의 상관관계를 계산

+ 파일의 데이터를 읽어 데이터프레임 만들기
 + read_csv('파일명') : CSV 파일 읽기 
 + read_excel('파일명') : 엑셀 파일 읽기
+ 파일로 저장하기 
  + to_csv('파일명') : CSV파일로 저장
  + to_excel('파일명') : 엑셀 파일로 저장

###데이터 간 상관 관계

ex) 날씨가 더우면 아이스크림이 많이 팔릴까?, 위도와 기온의 관계는?

상관계수 : 두 데이터간의 관계를 나타내어주는 계수
+ 상관계수 > 0 : 한 값이 증가하면 다른 값도 증가 

=> 기온과 아이스크림의 판매량 -> 기온이 높아짐에 따라 아이스크림의 판매량도 늘어남

+ 상관계수 < 0 : 한 값이 증가하면 다른 값은 감소 

=>기온과 위도 -> 위도가 높아지면 기온은 낮아짐

+ 상관계수의 절댓갑이 1에 가까울수록 높은 상관성

+ df.corr() : 수치 데이터를 가진 열 간의 상관관계를 계산
