# 예보등급
+ 0~30 좋음 : 대기오염 관련 질환자군에서도 영향이 유발되지 않을 수준
+ 31~80 보통 : 환자군에게 만성 노출시 경미한 영향이 유발될 수 있는 수준
+ 81~150 나쁨 : 환자군 및 민감군에게 유해한 영향 유발, 일반인도 불쾌감 경험할 수준
+ 151이상 매우나쁨 : 환자군 및 민감군에게 급성 노출시 심각한 영향 유발, 일반인도 약학 영향 유발할 수준

```
dt = {'서울': 69, '부산': 83, '대구': 96, '인천': 79, '광주': 73, '대전': 71, '울산': 91, '세종': 58, '경기': 101}
dt2 = {'좋음': 0, '보통': 5, '나쁨': 4, '매우나쁨': 0}
```
<pre>
좋음 () : 대기오염 관련 질환자군에서도 영향이 유발되지 않을 수준
보통 (■■■■■) : 환자군에게 만성 노출시 경미한 영향이 유발될 수 있는 수준
나쁨 (■■■■) : 환자군 및 민감군에게 유해한 영향 유발, 일반인도 불쾌감 경험할 수준
매우나쁨 () : 환자군 및 민감군에게 급성 노출시 심각한 영향 유발, 일반인도 약학 영향 유발할 수준
</pre>

````python
data = ['서울', '69','부산', '83','대구','96','인천','79','광주','73','대전', '71','울산','91','세종','58', '경기','101']

dt = {}
dt2 = {'좋음':0, '보통':0, '나쁨':0, '매우나쁨':0}
dt3 = {'좋음':'대기오염 관련 질환자군에서도 영향이 유발되지 않을 수준', 
       '보통':'환자군에게 만성 노출시 경미한 영향이 유발될 수 있는 수준', 
       '나쁨':'환자군 및 민감군에게 유해한 영향 유발, 일반인도 불쾌감 경험할 수준', 
       '매우나쁨':'환자군 및 민감군에게 급성 노출시 심각한 영향 유발, 일반인도 약학 영향 유발할 수준'}
````

````python
for index in range(0,len(data) ,2):
  dt[data[index]] =  int(data[index +1])
````

````python
for k,v in dt.items():
  dt2k = ''
  if v <= 30 : dt2k = '좋음'
  elif v <= 80 : dt2k = '보통'
  elif v <= 150 : dt2k = '나쁨'
  else: dt2k = '매우나쁨'
  
  dt2[dt2k] = dt2[dt2k] + 1

dt2
````

````python
for k, v in dt2.items():
  print(f"{k} ({'■' * v}) : {dt3[k]}")
````
