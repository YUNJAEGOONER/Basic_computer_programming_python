# 문자열분리
다음 문자열은 에어코리아(https://www.airkorea.or.kr/web/sidoQualityCompare?itemCode=10008&pMENU_NO=102) 실시간 시도별 대기정보(PM2.5)자료이다.
문자열을 분리하여 지역과대기정보를 갖는 딕션너리를 작성하시오.

data = '서울,15,부산,12,대구,13,인천,24,광주,11,대전,11,울산,11,경기,16, 강원,10,충북,16,충남,16,전북,13,전남,9,경북,12,경남,10,제주,14,세종,4'

````python
data = '서울,15,부산,12,대구,13,인천,24,광주,11,대전,11,울산,11,경기,16, 강원,10,충북,16,충남,16,전북,13,전남,9,경북,12,경남,10,제주,14,세종,4'
````

````python
city = []
temp = []
str = ""

for i in range(0,len(data),1):
  if (data[i] != ","):
    str = str + data[i]
    if (i == len(data)-1):
      temp.append(str)
  else:
    if ( '0' <= str[0] and str[0] <= '9'):
      temp.append(str)
      str = ""
    else:
      city.append(str)
      str = ""

print(city)
print(temp)
````

````python
# 쉼표를 기준으로 문자열 자르기 

lt = data.split(",") 
print(lt)

mycity = []
mytemp = []

for i in range(0,len(lt),1):
  if i % 2 == 0:
    mycity.append(lt[i])
  else:
    mytemp.append(int(lt[i]))

print(mycity)
print(mytemp) 
````

````python
mydic = dict(zip(mycity,mytemp))
print(mydic)
````

````python

lt2 = data.split(",") 
print(lt2)

mycity2 = []
mytemp2 = []

for item in lt2:
  if item.isnumeric():
    mytemp2.append(int(item))
  else:
    mycity2.append(item)

print(mycity2)
print(mytemp2)

#avg
sum(mytemp2) % len(mytemp2)

#idex값 반환
print(mycity.index("대구"))
print(mytemp2[mycity.index("대구")])

# dic -> key - vlaue 
#zip -> 두 개의 리스트를 dictionary로 
mydic2 = dict(zip(mycity,mytemp))
print(mydic2)

# key값으로 해당 value 가져오기
mydic2["대구"]

#dic의 모든 키값 가져오기  
for key in mydic2:
  print(mydic2[key] , end="")

#key와 value가 순서쌍으로 보는 방법 
for item in mydic2.items():
  print(item)

for k,v in mydic2.items():
  print(k,v)
````

````python
#함수

x = '10'

def checkNum(x):
  # isnumeric -> 문자열 함수 
  if x.isnumeric():
    print('숫자')
    return int(x)
  else:
    print('문자')
    return -1

#call function 
print(checkNum(x) + checkNum('-5'))
````
