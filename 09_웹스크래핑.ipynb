````python
from urllib.request import urlopen 
````

````python
#ÎÑ§Ïù¥Î≤Ñ ÎåìÍ∏ÄÎßéÏùÄÎâ¥Ïä§
url = 'https://media.naver.com/press/214/ranking?type=comment'
print(type(url))
#Ïõπ ÏÑúÎ≤Ñ ÏöîÏ≤≠ ÏùëÎãµÍ∞ùÏ≤¥ Î∞òÌôò
print(urlopen(url))
# b : binary 
print("--- Binary --- ")
print(urlopen(url).read())
print("--- Decoded to utf8 --- ")
#print(urlopen(url).read().decode('utf8'))
data = urlopen(url).read().decode('utf8')
print("------------")
````

````python
#select : press_rainking_list classÏùò li
ul = html.select('.press_ranking_list > li')
#20Í∞úÏùò listÎ•º Í∞ÄÏ†∏Ïò¥
print(len(ul))
print(ul[1].select_one('.list_title'))
````

````python
#3) Beautiful Soup ÌååÏã± 
lis = html.select('.press_ranking_list > li')
print(type(lis), len(lis))

print(lis[0].select_one('.list_title'))

fline = ""

for li in lis:
  rank = li.select_one(".list_ranking_num").text
  title = li.select_one('.list_title').text.replace('"',"").replace("'","").replace(",","")
  num = int(li.select_one('.list_comment').text.replace("ÎåìÍ∏Ä","").replace("\n",""))
  gubun = li.select_one('.list_img').text.replace("\n","").count("ÎèôÏòÅÏÉÅ")
  
  if gubun == 0:
    gubun = "Í∏∞ÏÇ¨"
  else:
    gubun = "ÎèôÏòÅÏÉÅ"
  
  line = f"{rank},{title},{num},{gubun}\n"
  print(line)
  fline = fline + line
````
