## 1. 데이터를 추출하려는 웹사이트에 HTTP 요청 응답 자료 저장


### **urlopen() 함수**
> ``` 
from urllib.request import urlopen 
data = urlopen(url).read().decode('utf8')
```
* 웹 서버에 정보를 요청한 후, 돌려받은 응답을 저장하여 ‘응답 객체(HTTPResponse)’를 반환
* 반환된 응답 객체의 read() 메소드를 실행
 + 웹 서버가 응답한 데이터를 바이트 배열로 읽어들임
 + 읽어들인 바이트 배열
   - 이진수로 이루어진 수열이어서 텍스트 형식의 데이터를 decode() 메소드를 실행하여 문자열로 변환

````python
from urllib.request import urlopen 
````

````python
#네이버 댓글많은뉴스
url = 'https://media.naver.com/press/214/ranking?type=comment'
print(type(url))
#웹 서버 요청 응답객체 반환
print(urlopen(url))
# b : binary 
print("--- Binary --- ")
print(urlopen(url).read())
print("--- Decoded to utf8 --- ")
#print(urlopen(url).read().decode('utf8'))
data = urlopen(url).read().decode('utf8')
print("------------")
````

##2. 웹페이지를 분석해서 특정 패턴의 데이터를 가져옴




### Beautiful Soup 모듈

````python
#1) Beautiful Soup 모듈
from bs4 import BeautifulSoup as bs
````

### Beautiful Soup 객체 생성
> ```
bs = BeautifulSoup(markup, parser)
```
+ Beautiful Soup 파서 종류
  * html.parser : 파이썬 표준 라이브러리에 포함된 파서로 빠르지만 유연하지 않기 때문에 단순한 HTML문서에 사용
 + BeautifulSoup(markup, "html.parser") 

````python
#2) Beautiful Soup 객체 생성
html = bs(data,'html.parser')
````

###  Beautiful Soup 파싱
>+  Beautiful Soup 태그 파싱
  + .find(태그명) : 조건에 맞는 태그 1개만 찾음 
  + .find_all(태그명) : 조건에 맞는 모든 태그 찾음
+  Beautiful Soup CSS선택자 파싱
  + .select_one(선택자) : 조건에 맞는 CSS선택자 1개만 찾음 
  + .select(선택자) : 조건에 맞는 CSS선택자 찾음

참고) https://www.w3schools.com/


````python
t = html.find('title')
print(t)
````

````python
div = html.find_all('div')
print(div)
````

````python
ul = html.find_all('ul')
print(ul)
````

````python
#select : press_rainking_list class의 li
ul = html.select('.press_ranking_list > li')
#20개의 list를 가져옴
print(len(ul))
print(ul[1].select_one('.list_title'))
````

````python
#3) Beautiful Soup 파싱 
lis = html.select('.press_ranking_list > li')
print(type(lis), len(lis))

print(lis[0].select_one('.list_title'))

fline = ""

for li in lis:
  rank = li.select_one(".list_ranking_num").text
  title = li.select_one('.list_title').text.replace('"',"").replace("'","").replace(",","")
  num = int(li.select_one('.list_comment').text.replace("댓글","").replace("\n",""))
  gubun = li.select_one('.list_img').text.replace("\n","").count("동영상")
  
  if gubun == 0:
    gubun = "기사"
  else:
    gubun = "동영상"
  
  line = f"{rank},{title},{num},{gubun}\n"
  print(line)
  fline = fline + line
````
