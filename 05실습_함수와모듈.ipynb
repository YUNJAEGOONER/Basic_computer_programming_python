# **해결문제**

---

원의 반지름을 입력받아 아래 공식을 참고해서 원의 넓이와 둘레를 구하고, 아래와 같이 출력해주자.

---

* 공식
  + 원의 면적 : $A = \pi r^2$
  + 원의 둘레 : $C = 2\pi r$

```python
#math 모듈 추가
import math;

# 입력 : 원의 반지름  / 출력 : 원의 둘레와 넓이
# 함수 정의 def function_name (parameter)

def circleAC(radius):
  A = radius * radius * math.pi
  C = 2 * radius * math.pi
  
  return A, C
```

```python
#함수 호출 : parameter로 1 을 전달 ( radius = 1 )
circleAC(1)
```

```python
#input으로 반지름을 입력받아 입력받은 정수를 parameter로 전달
r = int(input("반지름을 입력하세요 :"))
#tuple 형태로 반환됨 -> 변수 A, C에 각각 return 값이 한 개 씩 저장
print(circleAC(r))
A , C = circleAC(r)
#소수 두 번째 자리 까지 나타내기 :2f
print(f"반지름 {r}인 원의 면적은 {A:.2f}, 둘레는 {C:.2f}입니다.")
```

# **해결문제**

**가위/바위/보 게임**
1.   사용자로부터 숫자를 입력받아 사용자의 가위/바위/보를 결정한다. (예: 가위 = 1, 바위 = 2, 보 = 3)
2.   random 모듈을 사용하여 1~3사이의 정수를 무작위로 생성하여 컴퓨터의 가위/바위/보를 결정한다.
3. 사용자와 컴퓨터의 결과를 비교하여 승패를 판단한다.
4. 5판 3선승제로 진행하며, 각 판이 끝날 때마다 현재 스코어를 출력한다.




1. 입력 : 사용자로부터 1, 2, 3을 입력 받는다. -> 정수 입력
2. 컴퓨터가 1,2,3 을 random 하게 만듬 -> random 모듈 사용 (import random)
3. 입력값과 컴퓨터가 랜덤으로 생성한 숫자 비교
4. 5판 3선 승제 -> 사용자 또는 컴퓨터 중 3승 먼저 달성시 프로그램 종료

```python
#random 모듈을 사용
import random
```

```python
def check(n):
  if n == 1:
    return "가위"
  elif n == 2 :
     return "바위"
  else:
    return "보"
```

```python
total = 0 #전체 횟수
win = 0 # user가 이긴횟수


for i in range(0,5,1):
  #사용자로부터 숫자를 입력
  user = int(input("가위 = 1, 바위 = 2, 보 = 3 입력"))
  
  #컴퓨터가 1에서 3사이의 랜덤한 값을 갖는다.
  computer = random.randint(1,3)


  total = total + 1
  #사용자와 컴퓨터의 결과를 비교
  if user == computer:
    print("비겼습니다.")
  elif (user == 1 and computer == 3) or (user == 2 and computer == 1) or (user == 3 and computer == 2):
    win = win + 1
    print("이겼습니다.")
  else:
    print("졌습니다")
  
  print(f"User : {check(user)}, Computer {check(computer)}")
  print(f"{total}번 {win}승")
  if win == 3:
    break
```

# **실습과제**


mm 단위의 길이를 입력하면 cm, m, inch, ft 등으로 단위가 변환되어 출력되는 함수를 정의하고 사용 예를 보이시오.

1. mm 단위의 길이와 변환하고자 하는 단위를 입력으로 받는 convertUnit 함수를 정의한다. def converUnit(len, unit)
2. 단위 환산은 아래와 같다
	+ 1mm ➝ 0.1cm
	+ 1mm ➝ 0.001m
	+ 1mm ➝ 0.03937inch
	+ 1mm ➝ 0.003281ft

```python
def converUnit(len, unit):
  if unit == "cm":
    len = len * 0.1
    return len, "cm" 
  elif unit == "m":
    len = len * 0.001
    return len , "m"  
  elif unit == "inch":
    len = len * 0.03937
    return len, "inch"
  elif unit == "ft":
    len = len * 0.003281
    return len, "ft"
```

```python
#mm의 단위와, 변환하고자 하는 단위를 입력받기
len = int(input("길이를 입력하세요(mm) : "))
unit = input("변환하고자 하는 단위를 입력하세요 :")
convertlen , convertunit = converUnit(len, unit)
print(f"{len}mm ➝ {convertlen}{convertunit}")
```
