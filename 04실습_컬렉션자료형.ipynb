# 해결문제

덧셈, 뺄셈, 곱셈, 나눗셈이 가능한 사칙연산 계산기를 만들어 보자. 

네 가지 연산자 중 하나와 두 실수를 입력하면 계산 결과를 확인할 수 있다. 

연산자를 입력하는 대신 ‘exit’을 입력하면 프로그램이 종료된다. 

주의할 점은 나눗셈 연산에서 0으로 나누는 연산은 허용하지 말아야 한다. 

````python
while True:
  op = input('연산자 입력(+,-,*,/, 종료:exit =>')
  if op == 'exit' :
    break
  elif op not in ['+','-','*','/']:
    print("입력오류")
    continue
  else:
    x = float(input("입력1=>"))
    y = float(input("입력2=>"))

    if op == '/':
      if y == 0:
        print("0으로 나눌 수 없습니다.")
        continue
      else:
        #나눗셈의 결과를 소수 두 번째 자리까지 나타내기
        print(f'{x} / {y} = {x/y:,.2f}')
    elif op == '-':
      print(f'{x} - {y} = {x-y}')
    elif op ==  '*':
      print(f'{x} x {y} = {x*y}')
    elif op == '+ ':
      print(f'{x} + {y} = {x+y}')
````

````python
#리스트 : [], 순서로 구분 
lt = [1, 2, 'A']

#튜플 : (), 순서로 구분
tp = (1, 2, 'A')

#딕션너리 : {}, 키로 구분
dt = {'k1':1, 'k3':'A', 'k2':2}


#자료 가져오기 
print(lt[1])
print(tp[1])
print(dt['k2'])
````

````python
#리스트, 튜플 자료 접근
for item in tp :
  print(item)
````

````python
#딕션너리 자료접근
for item in dt :
  print(item, dt[item])
````

````python
for item in dt.items() :
  print(item)
````

````python
for k, v in dt.items() :
  print(k, v)
````

````python
# dic = {}
# for i in range(0,len(country),1):
#   dic[country[i]] = capital[i]

dic = dict(zip(country, capital))

correct = 0;

for i in country:
  answer = input(f"{i}의 수도는 어디입니까? ")
  if (dic[i] == answer):
    correct = correct + 1
    print("정답입니다.")
  else:
    print("오답입니다.")

print(f"총 {len(dic)} 중 정답수는 {correct}입니다.")
````

# 실습과제

아래 그림과 같이 피타고라스 정리를 만족하는 삼각형의 변 길이에 대한 순서쌍 (a, b, c)를 구해 보자. 

삼각형 한 변의 길이는 1부터 30미만으로 정해보자.

+ a 변수는 1~30 까지 반복
+ b 변수는 a~30 까지 반복
+ c 변수는 1~30 까지 반복

![싈습문제5](https://drive.google.com/uc?export=view&id=1tB-1cw5XwTWD-KsQEbj9EhQZrvBUBfDM)

````python
pita_list = []

for a in range(1,30,1):
  for b in range(a,30,1):
    for c in range(1,30,1):
      if (a*a + b*b  == c*c):
        pita_list.append([a,b,c])

pita_list
````
