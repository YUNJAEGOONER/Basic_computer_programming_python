###데이터 시각화
-> 데이터가 가진 특징을 그래프와 같은 시각도구로 보여주는 것

인포그래픽 (information + graphics) : 데이터의 특징을 시각적으로 표현한 것으로, 정보를 빠르고 쉽게 전달하기 위해 사용

###matplotlib 
데이터를 정적이나 동적 이미지, 또는 대화형 방식으로 시각화할 수 는 기능 제공

pyplot 모듈 사용 ( import matplotlib.pyplot as plt )

+ figure() : 새 그림판 생성, 그림 번호 부여 및 크기 설정 (생략가능)
+ plot() : (x,y) 형태의 데이터를 이용하여 선그래프 그리기
  + 색상 : b (파란색), g (초록색), r (빨간색), c (청록색), y (노란색) ... 
  + 점 모양 : o (원), v (역삼각형), ^ (삼각형), s (네모) 
  + 선 유형 : - (solid line), -- (dashed line), -.(dash-dot line) ... 
  
  => plt.plot(temp, 'go--') : 초록색 / 원 / dashed line 
+ title() : 그래프 이름 설정하기
+ xlabel() : x축 이름 설정하기 
+ ylabel() : y축 이름 설정하기
+ legend() : 범례 설정하기 
+ show() : 그래프 이미지 보이기

축 간격 및 값 설정 
+ xticks() : x축 값 설정
+ yticks() : y축 값 설정

다양한 그래프 
+ plot() : 기본 선그래프
+ bar() : 막대그래프
+ scatter() : 산점도그래프

```python
temp = [10,20,40,30,90,100,70,50,60,80]

plt.title("그래프의 이름")
plt.xlabel("x축 이름")
plt.ylabel("y축 이름")

plt.plot(temp,'go--') #green / circle / dashed_line
plt.show()
```

```python
#plt.figure()
#figsize를 이용하여 가로, 세로 길이 조절 가능 (inch 단위)
plt.figure(figsize=(10,6))

plt.title("그래프의 이름")
plt.xlabel("x축 이름")
plt.ylabel("y축 이름")

plt.plot(temp,'r^--') #green / circle / dashed_line
plt.show()
```

```python
x = list(range(1,len(temp)+1))
plt.figure(figsize=(10,6))
plt.bar(x,temp,label = '막대그래프') # x 축 : 1,2, ... 10 / y축 : temp
plt.xticks(x,rotation = 90) # rotation = x축의 수치를 90도 회전
plt.show()
```

```python
#scatter : 산점도 그래프
xs = [i**2 for i in temp] #[100, 400, 1600, 900, 8100, 10000, 4900, 2500, 3600, 6400]
x = list (range(1,len(temp)+1))
plt.figure(figsize=(10,6))
plt.scatter(x,temp, s = xs, c = xs , label= '산점그래프') # x축 : 1,2...10 / y축 : temp / s = 점들의 사이즈 / c = 점들의 색  
plt.xticks(x, rotation = 90)

plt.show()
```

```python
#그래프 겹쳐 그리기

plt.figure(figsize=(10,6))
plt.plot(x,temp,'r--', label = '기본그래프')
plt.bar(x, temp, label = '막대그래프')
plt.scatter(x, temp, s = xs , c = xs , label = '산점도그래프')
plt.legend(loc = 'best') #best : 범례를 표현하기 위한 가장 적절한 위치에 범례를 나타냄
plt.show()
```

###seaborn 라이브러리
+ matplolib를 기반으로 한 확장 라이브러리 : import seaborn as sns
+ 그래프 종류
  + barplot() : 막대그래프
  + countplot() : 범주형 자료를 가지는 열의 값별 개수를 자동으로 계산
  + displot() : 분포와 밀도 확인
  + boxplot() : 상자그림(데이터 범위, 최대/최소, 사분위값 표시)
  + heatmap() : 수치형 자료를 값의 범위에 따른 색상으로 시각화
  + pairplot() : 수치형 자료 열들의 모든 쌍에 대해 관계 표시


```python
import seaborn as sns
```

```python
#barplot
plt.figure(figsize=(10,7))
sns.barplot(x='모델명', y = 'CO2배출량', data = df.head(10))
plt.title('CO2배출량현황')
plt.xticks(rotation=90)
plt.show()
```

```python
#countplot : 항목별 갯수를 세어주는 그래프
#hue : 값별로 개수를 셀(count) 자료열 

plt.figure(figsize=(10,7))
sns.countplot(y = '업체명', hue = '등급', data = df)
plt.title('업체별 현황')
plt.show()
```

```python
#distplot : 분포와 밀도 확인
#matplotlib의 hist 그래프와 kdeplot을 통합한 그래프이며, 분포와 밀도를 확인할 수 있음

plt.figure(figsize=(10,7))
sns.displot(df['연간유류비'])
plt.title('연간유류비 현황')
plt.show()
```

```python
#boxplot
#데이터의 분포를 확인하고 싶을 때 가장 많이 사용하는 시각화 방법 중 하나로 데이터의 '범위'를 보여주고, 데이터의 상당 부분이 어디에 있는지, '이상치'가 있는지 확인할 수 있는 그래프
#다섯 수치 요약을 가지고 표현하는 그래프
#최솟값 : 제 1사분위에서 1.5 IQR1을 뺀 위치
#제 1사분위(Q1) : 25%의 위치를 의미
#제 2사분위(Q2) : 50%의 위치로 중앙값(median)을 의미
#제 3사분위(Q3) : 75%의 위치를 의미
#최댓값 : 제 3 사분위에서 1.5IQR을 더한 위치
#이상치(Outlier):최솟값과 최댓값을 넘어가는 위치에 있는 값

plt.figure(figsize=(10,8))
sns.boxplot(x='등급', y='연간유류비',hue='구분', data = df)
plt.show()
```

```python
#heatmap : 수치 데이터를 값의 범위에 따른 색상을 이용해 시각화
#다양한 값을 갖는 숫자 데이터를 열분포 형태와 같이 색상을 이용해서 시각화
#annot: 값 표시 (상자마다 값이 나타남), cbar : 색상바 표시

plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot = True, cbar = True)
plt.show()
```

```python
#pairplot : 서로 다른 데이터쌍 : 산점도 그래프
#같은 데이터가 만나는 대각선 : 히스토그램
#수치형 자료를 가지는 열들을 그리도(grid)형태로 각 데이터 열의 조합에 대해 산점도 그래프를 그리고 같은 데이터가 만나는 대각선 영역에는 해당 데이터의 히스토그램을 그림

sns.pairplot(data=df,hue='구분') # 구분 (국산차와 수입차)
plt.show()
```
